```python
import matplotlib as mpl
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
# intro to a basic plot 
x= np.linspace(0,10,50)
plt.plot(x,np.tan(x), '--')
plt.plot(x, np.cos(x)), '..'
plt.show()

```

```python
#creating a subplot
plt.subplot(2,1,1) #(rows, columns, panel number)
plt.plot(x,np.sin(x), '-')

plt.subplot(2,1,2)
plt.plot(x,np.cos(x),'--')
```

```python
#plotting graphs with colors
plt.plot(x, np.sin(x - 0), color='red')    
plt.plot(x, np.sin(x - 1), color='c')       
plt.plot(x, np.sin(x - 2), color='0.55')
```

```python
#introducing axis limits
plt.plot(x,np.cos(x))
plt.xlim(-5,10)
plt.ylim(-2,5)
```

```python
plt.plot(x, np.sin(x))
plt.axis('tight');
```

```python
plt.plot(x, np.sin(x))
plt.axis('equal')
```

```python
# labelling a plot and axes
axis = plt.axes()
axis.plot(x, np.cos(x), '--')
axis.set(xlim=(0, 10), ylim=(-4, 4),
xlabel='x', ylabel='cos(x)',
title='matplotlib plot');
```

```python
#using a scatter plot
x=np.linspace(1,20,40)
y=np.sin(x)
plt.plot(x,y,'o',color='blue')
```

```python
plt.scatter(x,y,marker ='o')
```

```python
rng = np.random.RandomState(0)
x = rng.randn(100)
y = rng.randn(100)
colors = rng.rand(100)
sizes = 100 * rng.rand(100)
plt.scatter(x, y, c=colors, s=sizes, alpha=0.4, cmap='viridis')
plt.colorbar();  # show color scale
```

```python
A = [15., 30., 45., 22.]

B = [15., 25., 50., 20.]

z2 = range(4)

plt.bar(z2, A, color = 'blue')
plt.bar(z2, B, color = 'red', bottom = A)

plt.show()
```
